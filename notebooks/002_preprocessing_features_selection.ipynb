```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carica il dataset bilanciato
df = pd.read_csv("../data/prodromi_ctgan_balanced_clean.csv")
target_col = "prodromi_combo"

# 2. Escludi i 4 prodromi e il target
exclude_cols = ['Hyposmia', 'Constipation', 'REM', 'Depression', target_col]
X = df.drop(columns=exclude_cols)
y = df[target_col]

# 3. Tieni solo le variabili numeriche
X = X.select_dtypes(include=np.number)

# 4. Imputazione NaN
imputer = SimpleImputer(strategy="mean")
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# 5. Standardizzazione
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_imputed), columns=X.columns)

# ================================
# 🎯 1. FILTER METHOD (F-test)
# ================================
filter_selector = SelectKBest(score_func=f_classif, k='all')
filter_selector.fit(X_scaled, y)
filter_scores = pd.DataFrame({
    "Feature": X.columns,
    "F_score": filter_selector.scores_,
    "p_value": filter_selector.pvalues_
}).sort_values(by="F_score", ascending=False)

# ================================
# 🔁 2. WRAPPER METHOD (RFE)
# ================================
log_clf = LogisticRegression(max_iter=1000, solver='liblinear')
wrapper_selector = RFE(log_clf, n_features_to_select=10)
wrapper_selector.fit(X_scaled, y)
wrapper_selected = X.columns[wrapper_selector.support_]

# ================================
# 🌲 3. EMBEDDED METHOD (Random Forest)
# ================================
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_scaled, y)
rf_importances = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_clf.feature_importances_
}).sort_values(by="Importance", ascending=False)

# ================================
# 📊 Visualizzazioni
# ================================
plt.figure(figsize=(10,6))
sns.barplot(data=filter_scores.head(10), x='F_score', y='Feature', palette='crest')
plt.title("🔬 Filter Method - Top 10 F-scores")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(data=rf_importances.head(10), x='Importance', y='Feature', palette='viridis')
plt.title("🌲 Embedded Method - Random Forest Importances")
plt.tight_layout()
plt.show()

# ================================
# ✅ Risultati testuali
# ================================
print("\n✅ Wrapper Method - Features selezionate da RFE:")
print(wrapper_selected.tolist())

print("\n✅ Filter Method - Features con p < 0.05:")
print(filter_scores[filter_scores['p_value'] < 0.05]['Feature'].tolist())

print("\n✅ Embedded Method - Top 10 feature da Random Forest:")
print(rf_importances.head(10)['Feature'].tolist())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Carica il dataset completo
df = pd.read_csv("../data/prodromi_ctgan_balanced_clean.csv")

# 2. Target
target_col = "prodromi_combo"

# 3. Features selezionate da almeno due metodi
selected_features = [
    '%det ML',
    '%det AP',
    'Sex (M=1, F=2)',
    'Age',
    'Heigth',
    'Weigth',
    'Onset (1=early (<49), 2= middle (5 2-69), 3=late (>7 2) )',
    'Postural Alteration\n(1=si,2=no)',
    'Double Support',
    'Stride Length',
    'Single Support',
    'Duration (years)',
    'Cadence',
    'Tilt'
]

# 4. Subset del dataframe
df_subset = df[selected_features + [target_col]]

# 5. Calcola la matrice di correlazione assoluta tra le feature (escludendo il target)
corr_matrix = df_subset[selected_features].corr().abs()

# 6. Trova le coppie con correlazione > 0.85
high_corr = (
    corr_matrix.where((corr_matrix > 0.85) & (corr_matrix < 1.0))
    .stack()
    .reset_index()
)
high_corr.columns = ['Feature 1', 'Feature 2', 'Correlation']
high_corr = high_corr.sort_values(by='Correlation', ascending=False)

# 7. Mostra la heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Heatmap delle correlazioni tra le feature selezionate")
plt.tight_layout()
plt.show()

# 8. Stampa le coppie altamente correlate
print("🔍 Coppie con correlazione > 0.85:")
print(high_corr)
```

```python
final_features = [
    '%det ML',
    '%det AP',
    'Sex (M=1, F=2)',
    'Age',
    'Onset (1=early (<49), 2= middle (5 2-69), 3=late (>7 2) )',
    'Postural Alteration\n(1=si,2=no)',
    'Double Support',
    'Stride Length',
    'Single Support',
    'Duration (years)',
    'Cadence',
    'Tilt'
]

df_final = df[final_features + [target_col]]

# 🔽 Salva il nuovo dataset
df_final.to_csv("prodromi_ctgan_final.csv", index=False)
print("✅ Dataset finale salvato come 'prodromi_ctgan_final.csv'")
```
